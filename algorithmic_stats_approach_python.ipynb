```python
import pandas as pd
```

```python
data_frame = pd.read_excel('/home/ashish/projects/tukeyhsd_2_tensileStrengthMeans/cotton weight.xlsx')
data_frame
```

```python
id_vars = ['index']
value_vars = ['cotwt.15','cotwt.20','cotwt.25','cotwt.30','cotwt.35']
data_new = pd.melt(data_frame.reset_index(),id_vars,value_vars)
data_new.columns = ['id','treatments','values']
data_new
```

```python
from statsmodels.stats.multicomp import MultiComparison
from statsmodels.stats.multicomp import tukeyhsd
```

```python
mc = MultiComparison(data_new['values'],data_new['treatments'])
mc_result = mc.tukeyhsd(0.05)
mc_result.summary()
```

## So from the above output table we can easily determine which pair of means are significantly different OR same.
